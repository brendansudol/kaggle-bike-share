```python
import itertools
from time import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import StratifiedKFold, train_test_split
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import mean_squared_error
```

```python
# load in datasets
dfs = {}
for name in ['train', 'test']:
    df = pd.read_csv('../data/%s.csv' % name)
    df['_data'] = name
    dfs[name] = df

# combine train and test data into one df
df = dfs['train'].append(dfs['test'])

# lowercase column names
df.columns = map(str.lower, df.columns)

df.head()
```

```python
# add some date/time related columns

dt = pd.DatetimeIndex(df['datetime'])

df['date'] = dt.date
df['day'] = dt.day
df['month'] = dt.month
df['quarter'] = dt.quarter
df['year'] = dt.year
df['hour'] = dt.hour
df['dow'] = dt.dayofweek
df['doy'] = dt.dayofyear
df['woy'] = dt.weekofyear

# normalize year col (2011 = 0, 2012 = 1)
min_year = df['year'].min()
df['year_norm'] = df['year'].apply(lambda x: x - min_year)

# apply logarithmic transformation of dependent cols
# (adding 1 first so that 0 values don't become -inf)
for col in ['casual', 'registered', 'count']:
    df['%s_log' % col] = np.log(df[col] + 1)

df.iloc[0]
```

```python
def get_data():
    data = df[df['_data'] == 'train'].copy()
    return data


def custom_train_test_split(data, cutoff_day=10):
    train = data[data['day'] <= cutoff_day]
    test = data[data['day'] > cutoff_day]
    
    return train, test


def prep_data(data, input_cols):
    X = data[input_cols].as_matrix()
    y_r = data['registered_log'].as_matrix()
    y_c = data['casual_log'].as_matrix()
    
    return X, y_r, y_c


def get_rmsle(y_pred, y_actual):
    return np.sqrt(np.square(np.log(y_pred + 1) - np.log(y_actual + 1)).mean())

    
def predict(input_cols, model_params={}):
    data = get_data()
    
    train, test = custom_train_test_split(data)
    
    X_train, y_train_r, y_train_c = prep_data(train, input_cols)
    X_test, y_test_r, y_test_c = prep_data(test, input_cols)

    model_params.update({
        'n_jobs': -1,
        'random_state': 123,
    })
    model = RandomForestRegressor(**model_params)
    
    model_r = model.fit(X_train, y_train_r)
    y_pred_r = np.exp(model_r.predict(X_test)) - 1
    
    model_c = model.fit(X_train, y_train_c)
    y_pred_c = np.exp(model_c.predict(X_test)) - 1

    y_pred_comb = np.round(y_pred_r + y_pred_c)
    y_pred_comb[y_pred_comb < 0] = 0
    
    y_test_comb = np.exp(y_test_r) + np.exp(y_test_c) - 2

    score = get_rmsle(y_pred_comb, y_test_comb)
    return score
```

```python
x_cols = ['dow', 'woy', 'hour', 'year_norm', 'season', 'holiday', 'workingday', 
          'weather', 'temp', 'atemp', 'humidity', 'windspeed', 'day']

predict(x_cols, {'n_estimators': 500})
```

```python
test_cols = ['dow', 'woy', 'temp', 'atemp', 'season']

col_combos = []
for x in range(3, len(test_cols) + 1):
    for cols in itertools.combinations(test_cols, x):
        col_combos.append(list(cols))
total_combos = len(col_combos)
print 'Total feature combinations: %d' % total_combos


t0 = time()
scores = []
for i, cols in enumerate(col_combos):
    score = predict(cols)
    scores.append((cols, score))
    
    # output progress (on roughly 10% intervals)
    j, milestone = i + 1, max(1, total_combos / 10)
    if j % milestone == 0 or j == total_combos:
        print '{} of {} done ({} secs)'.format(
            j, total_combos, round(time() - t0, 1),
        )
```

```python
sorted_scores = sorted(scores, reverse=True, key=lambda x: -x[1])

print 'top models:\n'
for i, score in enumerate(sorted_scores[:5]):
    print 'rank: %d\nfeatures: %s\nscore: %f\n' % (i + 1, score[0], score[1])
```
